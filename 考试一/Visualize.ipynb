```python
import pandas as pd
import seaborn as sns; sns.set_theme()
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 300
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
nodes = pd.read_csv("nodes.csv")
nodes.head()
```

```python
counts = nodes['modularity_class'].value_counts()
plt.pie(counts, labels = counts.index, explode=[0.02 for _ in range(len(counts))], autopct='%1.1f%%')
plt.title("不同社团比重")
plt.show()
```

```python
sns.displot(nodes, x="closnesscentrality")
plt.title("接近中心性分布")
plt.xlabel("Clossness Centrality")
plt.show()
```

```python
sns.displot(nodes, x="betweenesscentrality")
plt.title("介数中心性分布")
plt.xlabel("Betweeness Centrality")
plt.show()
```

```python
sns.displot(nodes, x="pageranks")
plt.title("PageRank分布")
plt.xlabel("PageRank")
plt.show()
```

```python
sns.displot(nodes, x="eigencentrality")
plt.title("特征向量中心性分布")
plt.xlabel("Eigenvector Centrality")
plt.show()
```
